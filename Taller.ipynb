**1. Se desea estimar la altura promedio de estudiantes universitarios. En una muestra de 50 estudiantes,
se obtuvo una media de 170 cm. Se sabe que la desviación estándar poblacional es de 10 cm.
Construye un intervalo de confianza del 95 % para la media poblacional.**

````python
media_muestral = 170
tamano_muestra = 50
desviacion_estandar_poblacional = 10
nivel_confianza = 0.95
````

## Calcular el valor crítico (z-score)

### Subtask:
Determina el valor de z para el nivel de confianza del 95%.


````python
from scipy.stats import norm

alpha = 1 - nivel_confianza
valor_critico = norm.ppf(1 - alpha / 2)
````

````python
import numpy as np

error_estandar = desviacion_estandar_poblacional / np.sqrt(tamano_muestra)
````

````python
margen_error = valor_critico * error_estandar
````

````python
limite_inferior = media_muestral - margen_error
limite_superior = media_muestral + margen_error

print(f"El intervalo de confianza del 95% para la altura promedio es: ({limite_inferior:.2f}, {limite_superior:.2f}) cm")
````

**11. El departamento de salud planea una encuesta para estimar la proporción de residentes vacunados
contra la influenza. En un estudio piloto, el 72 % de una muestra mostró estar vacunado. Determine
el tamaño de muestra necesario para estimar esta proporción con un margen de error del 3 % y un
nivel de confianza del 95 %. Si no se tuviera el estudio piloto, ¿cuál sería el tamaño requerido?**

````python
import numpy as np
from scipy.stats import norm

# Escenario 1: Con estudio piloto
# 1. Define los parámetros para el primer escenario
proporcion_muestral = 0.72
margen_error = 0.03
nivel_confianza = 0.95

# 2. Calcular el valor crítico (z-score)
alpha = 1 - nivel_confianza
valor_critico_z = norm.ppf(1 - alpha / 2)

# 3. Calcular el tamaño de muestra (con estudio piloto)
# Fórmula: n = (Z^2 * p * (1-p)) / E^2
tamano_muestra_piloto = (valor_critico_z**2 * proporcion_muestral * (1 - proporcion_muestral)) / margen_error**2
tamano_muestra_piloto = np.ceil(tamano_muestra_piloto) # Redondear hacia arriba

# 4. Mostrar el resultado (con estudio piloto)
print(f"Con estudio piloto (proporción muestral = {proporcion_muestral}):")
print(f"El tamaño de muestra necesario es: {int(tamano_muestra_piloto)}")

print("-" * 30) # Separador

# Escenario 2: Sin estudio piloto
# 5. Define los parámetros para el segundo escenario
# Margen de error y nivel de confianza son los mismos
# Asumimos la proporción más conservadora p = 0.5
proporcion_conservadora = 0.5

# El valor crítico z es el mismo ya que el nivel de confianza es el mismo

# 6. Calcular el tamaño de muestra (sin estudio piloto)
# Fórmula: n = (Z^2 * 0.25) / E^2  (ya que p*(1-p) es máximo cuando p=0.5, resultando en 0.25)
tamano_muestra_sin_piloto = (valor_critico_z**2 * proporcion_conservadora * (1 - proporcion_conservadora)) / margen_error**2
tamano_muestra_sin_piloto = np.ceil(tamano_muestra_sin_piloto) # Redondear hacia arriba

# 7. Mostrar el resultado (sin estudio piloto)
print(f"Sin estudio piloto (proporción asumida = {proporcion_conservadora}):")
print(f"El tamaño de muestra necesario es: {int(tamano_muestra_sin_piloto)}")
````

**Justificación del uso de p = 0.5 cuando no hay datos preliminares:**

En el cálculo del tamaño de muestra para una proporción, la fórmula incluye el término `p * (1-p)`, donde `p` es la proporción poblacional que se desea estimar. Este término representa la variabilidad de la proporción.

La función `p * (1-p)` alcanza su valor máximo cuando `p = 0.5`. En este punto, `p * (1-p) = 0.5 * (1-0.5) = 0.25`.

Cuando no se tiene información preliminar sobre la proporción poblacional (es decir, no hay un estudio piloto, encuesta previa, o datos históricos confiables), se utiliza `p = 0.5` para calcular el tamaño de muestra. Esto se hace porque, al usar la proporción que maximiza la variabilidad (`p = 0.5`), se garantiza el tamaño de muestra más grande posible para un margen de error y nivel de confianza dados.

Este enfoque conservador asegura que el tamaño de muestra calculado será suficiente para lograr el margen de error deseado, independientemente de cuál sea la verdadera proporción poblacional. Si se utilizara una proporción diferente sin tener datos que la respalden, el tamaño de muestra resultante podría ser insuficiente para alcanzar la precisión requerida si la verdadera proporción estuviera más cerca de 0.5.

**21. Una empresa afirma que sus paquetes de cereal tienen un peso promedio de 500 g. Se sospecha que
la máquina está descalibrada y subllenando. Se toma una muestra de 40 paquetes, obteniendo una
media de 495 g. La desviación estándar poblacional es σ = 15 g. Con un nivel de de significancia**


```
α = 0,05.
1. Plantee las hipótesis nula y alternativa.
2. Calcule el estadístico Z.
3. Determine la región crítica y la región de rechazo.
4. Concluya si se rechaza H0.**
```




````python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# 1. Definir las hipótesis
# H0: La media poblacional es igual a 500 g (μ = 500)
# H1: La media poblacional es menor que 500 g (μ < 500) - Prueba unilateral izquierda

# 2. Establecer el nivel de significancia α
nivel_significancia = 0.05

# Datos del problema
media_muestral = 495
media_poblacional_hipotetizada = 500
desviacion_estandar_poblacional = 15
tamano_muestra = 40

# 3. Identificar el estadístico de prueba y su distribución (Z y distribución normal estándar)

# 4. Calcular el estadístico de prueba Z
error_estandar = desviacion_estandar_poblacional / np.sqrt(tamano_muestra)
estadistico_z = (media_muestral - media_poblacional_hipotetizada) / error_estandar

print(f"Estadístico Z calculado: {estadistico_z:.4f}")

# 5. Determinar el valor crítico y la región de rechazo
# Para una prueba unilateral izquierda, el valor crítico es el valor de Z tal que P(Z <= valor_critico) = alpha
valor_critico_z = norm.ppf(nivel_significancia)

print(f"Valor crítico Z para α = {nivel_significancia}: {valor_critico_z:.4f}")

# Región de rechazo: Z < valor_critico_z

# 6. Tomar una decisión estadística
print("\nDecisión estadística:")
if estadistico_z < valor_critico_z:
    print(f"El estadístico Z calculado ({estadistico_z:.4f}) es menor que el valor crítico Z ({valor_critico_z:.4f}).")
    print("Se rechaza la hipótesis nula (H0).")
else:
    print(f"El estadístico Z calculado ({estadistico_z:.4f}) no es menor que el valor crítico Z ({valor_critico_z:.4f}).")
    print("No se rechaza la hipótesis nula (H0).")

# 7. Calcular el valor p
# Para una prueba unilateral izquierda, el valor p es P(Z <= estadistico_z)
valor_p = norm.cdf(estadistico_z)

print(f"\nValor p: {valor_p:.4f}")

# Conclusión basada en el valor p
print("Decisión basada en el valor p:")
if valor_p < nivel_significancia:
    print(f"El valor p ({valor_p:.4f}) es menor que el nivel de significancia ({nivel_significancia}).")
    print("Se rechaza la hipótesis nula (H0).")
else:
    print(f"El valor p ({valor_p:.4f}) no es menor que el nivel de significancia ({nivel_significancia}).")
    print("No se rechaza la hipótesis nula (H0).")


# 8. Graficar la distribución y la región de rechazo
plt.figure(figsize=(10, 6))
x = np.linspace(-4, 4, 1000)
plt.plot(x, norm.pdf(x, 0, 1), label='Distribución Normal Estándar')

# Región de rechazo
x_rechazo = np.linspace(-4, valor_critico_z, 100)
plt.fill_between(x_rechazo, 0, norm.pdf(x_rechazo, 0, 1), color='red', alpha=0.5, label='Región de Rechazo (α)')

# Valor crítico
plt.axvline(valor_critico_z, color='red', linestyle='--', label=f'Valor Crítico Z ({valor_critico_z:.2f})')

# Estadístico Z calculado
plt.axvline(estadistico_z, color='blue', linestyle='-', label=f'Estadístico Z Calculado ({estadistico_z:.2f})')

plt.title('Distribución Normal Estándar con Región de Rechazo (Prueba Unilateral Izquierda)')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# 9. Concluir en términos prácticos
print("\nConclusión en términos prácticos:")
if valor_p < nivel_significancia:
    print(f"Con un nivel de significancia del {nivel_significancia*100:.0f}%, hay suficiente evidencia estadística para concluir que el peso promedio de los paquetes de cereal es significativamente menor a 500 g.")
    print("Esto sugiere que la máquina podría estar descalibrada y subllenando los paquetes.")
else:
    print(f"Con un nivel de significancia del {nivel_significancia*100:.0f}%, no hay suficiente evidencia estadística para concluir que el peso promedio de los paquetes de cereal es significativamente menor a 500 g.")
    print("No se puede afirmar con esta muestra que la máquina esté descalibrada y subllenando.")

print(f"\nEl valor p de {valor_p:.4f} representa la probabilidad de obtener una media muestral de 495 g (o una más extrema) si la verdadera media poblacional fuera 500 g. Dado que este valor es menor que α = {nivel_significancia}, consideramos que este resultado es improbable si la hipótesis nula fuera cierta, lo que nos lleva a rechazar H0.")
````

**23. Un profesor afirma que el promedio de su clase en un examen nacional es diferente al promedio nacional de 75 puntos. Una muestra de 36 estudiantes de su clase tiene una media de 78. La desviación estándar poblacional es σ = 12. Con un nivel de significancia: α = 0,05.**


```
1. Plantee las hipótesis nula y alternativa.
2. Calcule el estadístico Z.
3. Determine la región crítica y la región de rechazo.
4. Concluya si se rechaza H0.
```




````python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Problema: Prueba de hipótesis sobre la media de un examen nacional

# 1. Plantear las hipótesis nula y alternativa.
# El profesor afirma que el promedio es DIFERENTE al promedio nacional (75 puntos).
# Como la afirmación es que es DIFERENTE (puede ser mayor o menor), es una prueba bilateral.
# H0: La media poblacional de la clase es igual al promedio nacional (μ = 75)
# H1: La media poblacional de la clase es diferente al promedio nacional (μ ≠ 75) - Prueba bilateral

print("1. Hipótesis planteadas:")
print("H0: μ = 75")
print("H1: μ ≠ 75")
print("-" * 30)

# 2. Establecer el nivel de significancia α.
nivel_significancia = 0.05
print(f"2. Nivel de significancia α: {nivel_significancia}")
print("-" * 30)

# Datos del problema
media_muestral = 78
media_poblacional_hipotetizada = 75
desviacion_estandar_poblacional = 12
tamano_muestra = 36

# 3. Identificar el estadístico de prueba (Z, t, χ2) y su distribución.
# Dado que se conoce la desviación estándar poblacional (σ) y el tamaño de la muestra es grande (n >= 30),
# utilizamos el estadístico Z y la distribución normal estándar.
print("3. Estadístico de prueba: Z")
print("Distribución: Normal estándar")
print("-" * 30)

# 4. Determinar el valor crítico y la región de rechazo.
# Para una prueba bilateral con nivel de significancia α, hay dos valores críticos:
# uno negativo (-Z_alpha/2) y uno positivo (+Z_alpha/2).
# El área en cada cola de la distribución es α/2.
alpha_dos_colas = nivel_significancia / 2
valor_critico_z_inferior = norm.ppf(alpha_dos_colas) # Valor crítico izquierdo
valor_critico_z_superior = norm.ppf(1 - alpha_dos_colas) # Valor crítico derecho

print(f"4. Valores críticos Z para α = {nivel_significancia}: {valor_critico_z_inferior:.4f} y {valor_critico_z_superior:.4f}")
print(f"Región de rechazo: Z < {valor_critico_z_inferior:.4f} o Z > {valor_critico_z_superior:.4f}")
print("-" * 30)

# 5. Calcule el estadístico de prueba y el valor p.
error_estandar = desviacion_estandar_poblacional / np.sqrt(tamano_muestra)
estadistico_z = (media_muestral - media_poblacional_hipotetizada) / error_estandar

print(f"5. Estadístico Z calculado: {estadistico_z:.4f}")

# Para una prueba bilateral, el valor p es el doble del área en la cola más allá del estadístico Z calculado.
# Dado que el estadístico Z es positivo en este caso (78 > 75), calculamos el área a la derecha y la multiplicamos por 2.
# Si el estadístico Z fuera negativo, calcularíamos el área a la izquierda y la multiplicaríamos por 2.
valor_p = 2 * (1 - norm.cdf(abs(estadistico_z))) # Usamos abs() para manejar casos con Z negativo

print(f"Valor p: {valor_p:.4f}")
print("-" * 30)

# 6. Graficar la distribución correspondiente, indicando claramente la región de rechazo, el valor crítico y el estadístico calculado.
plt.figure(figsize=(10, 6))
x = np.linspace(-4, 4, 1000)
plt.plot(x, norm.pdf(x, 0, 1), label='Distribución Normal Estándar')

# Región de rechazo (dos colas)
x_rechazo_inferior = np.linspace(-4, valor_critico_z_inferior, 100)
plt.fill_between(x_rechazo_inferior, 0, norm.pdf(x_rechazo_inferior, 0, 1), color='red', alpha=0.5, label=f'Región de Rechazo (α/2)')

x_rechazo_superior = np.linspace(valor_critico_z_superior, 4, 100)
plt.fill_between(x_rechazo_superior, 0, norm.pdf(x_rechazo_superior, 0, 1), color='red', alpha=0.5) # No repetir la etiqueta

# Valores críticos
plt.axvline(valor_critico_z_inferior, color='red', linestyle='--', label=f'Valores Críticos Z ({valor_critico_z_inferior:.2f}, {valor_critico_z_superior:.2f})')
plt.axvline(valor_critico_z_superior, color='red', linestyle='--') # No repetir la etiqueta

# Estadístico Z calculado
plt.axvline(estadistico_z, color='blue', linestyle='-', label=f'Estadístico Z Calculado ({estadistico_z:.2f})')

plt.title('Distribución Normal Estándar con Región de Rechazo (Prueba Bilateral)')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
print("-" * 30)

# 7. Tomar una decisión estadística y concluir en términos prácticos, explicando el significado del valor p.
print("7. Decisión estadística y conclusión:")

# Decisión basada en la región de rechazo
if estadistico_z < valor_critico_z_inferior or estadistico_z > valor_critico_z_superior:
    print(f"El estadístico Z calculado ({estadistico_z:.4f}) cae en la región de rechazo.")
    print("Se rechaza la hipótesis nula (H0).")
else:
    print(f"El estadístico Z calculado ({estadistico_z:.4f}) no cae en la región de rechazo.")
    print("No se rechaza la hipótesis nula (H0).")

# Decisión basada en el valor p
print("\nDecisión basada en el valor p:")
if valor_p < nivel_significancia:
    print(f"El valor p ({valor_p:.4f}) es menor que el nivel de significancia ({nivel_significancia}).")
    print("Se rechaza la hipótesis nula (H0).")
else:
    print(f"El valor p ({valor_p:.4f}) no es menor que el nivel de significancia ({nivel_significancia}).")
    print("No se rechaza la hipótesis nula (H0).")

# Concluir en términos prácticos
print("\nConclusión en términos prácticos:")
if valor_p < nivel_significancia:
    print(f"Con un nivel de significancia del {nivel_significancia*100:.0f}%, hay suficiente evidencia estadística para concluir que la media de los estudiantes del profesor en el examen nacional es significativamente diferente al promedio nacional de 75 puntos.")
    print("Esto respalda la afirmación del profesor.")
else:
    print(f"Con un nivel de significancia del {nivel_significancia*100:.0f}%, no hay suficiente evidencia estadística para concluir que la media de los estudiantes del profesor en el examen nacional es significativamente diferente al promedio nacional de 75 puntos.")
    print("No se puede afirmar con esta muestra que el promedio de la clase sea diferente al promedio nacional.")

print(f"\nEl valor p de {valor_p:.4f} representa la probabilidad de obtener una media muestral de 78 puntos (o una tan o más extrema en cualquier dirección) si la verdadera media poblacional de la clase fuera 75 puntos. Dado que este valor es mayor que α = {nivel_significancia}, consideramos que este resultado no es lo suficientemente improbable si la hipótesis nula fuera cierta, lo que nos lleva a no rechazar H0.")
````

````python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Problema: Prueba de hipótesis sobre el llenado de botellas

# 1. Plantear las hipótesis nula y alternativa.
# La afirmación implícita de la máquina es que llena las botellas a 500 mL. La auditoría sugiere que podría ser diferente (o menor, aunque no se especifica la dirección).
# Como no se especifica si el interés es solo si es menor o mayor, asumimos una prueba bilateral (diferente).
# H0: La media poblacional del volumen de llenado es igual a 500 mL (μ = 500)
# H1: La media poblacional del volumen de llenado es diferente de 500 mL (μ ≠ 500) - Prueba bilateral

print("1. Hipótesis planteadas:")
print("H0: μ = 500")
print("H1: μ ≠ 500")
print("-" * 30)

# 2. Establecer el nivel de significancia α.
nivel_significancia = 0.01
print(f"2. Nivel de significancia α: {nivel_significancia}")
print("-" * 30)

# Datos del problema
media_muestral = 498
media_poblacional_hipotetizada = 500
desviacion_estandar_poblacional = 5
tamano_muestra = 60

# 3. Identificar el estadístico de prueba (Z, t, χ2) y su distribución.
# Dado que se conoce la desviación estándar poblacional (σ) y el tamaño de la muestra es grande (n >= 30),
# utilizamos el estadístico Z y la distribución normal estándar.
print("3. Estadístico de prueba: Z")
print("Distribución: Normal estándar")
print("-" * 30)

# 4. Determinar el valor crítico y la región de rechazo.
# Para una prueba bilateral con nivel de significancia α, hay dos valores críticos:
# uno negativo (-Z_alpha/2) y uno positivo (+Z_alpha/2).
# El área en cada cola de la distribución es α/2.
alpha_dos_colas = nivel_significancia / 2
valor_critico_z_inferior = norm.ppf(alpha_dos_colas) # Valor crítico izquierdo
valor_critico_z_superior = norm.ppf(1 - alpha_dos_colas) # Valor crítico derecho

print(f"4. Valores críticos Z para α = {nivel_significancia}: {valor_critico_z_inferior:.4f} y {valor_critico_z_superior:.4f}")
print(f"Región de rechazo: Z < {valor_critico_z_inferior:.4f} o Z > {valor_critico_z_superior:.4f}")
print("-" * 30)

# 5. Calcule el estadístico de prueba y el valor p.
error_estandar = desviacion_estandar_poblacional / np.sqrt(tamano_muestra)
estadistico_z = (media_muestral - media_poblacional_hipotetizada) / error_estandar

print(f"5. Estadístico Z calculado: {estadistico_z:.4f}")

# Para una prueba bilateral, el valor p es el doble del área en la cola más allá del estadístico Z calculado.
# Usamos abs() para manejar casos con Z positivo o negativo
valor_p = 2 * (1 - norm.cdf(abs(estadistico_z)))

print(f"Valor p: {valor_p:.4f}")
print("-" * 30)

# 6. Graficar la distribución correspondiente, indicando claramente la región de rechazo, el valor crítico y el estadístico calculado.
plt.figure(figsize=(10, 6))
x = np.linspace(-4, 4, 1000)
plt.plot(x, norm.pdf(x, 0, 1), label='Distribución Normal Estándar')

# Región de rechazo (dos colas)
x_rechazo_inferior = np.linspace(-4, valor_critico_z_inferior, 100)
plt.fill_between(x_rechazo_inferior, 0, norm.pdf(x_rechazo_inferior, 0, 1), color='red', alpha=0.5, label=f'Región de Rechazo (α/2 = {alpha_dos_colas})')

x_rechazo_superior = np.linspace(valor_critico_z_superior, 4, 100)
plt.fill_between(x_rechazo_superior, 0, norm.pdf(x_rechazo_superior, 0, 1), color='red', alpha=0.5) # No repetir la etiqueta

# Valores críticos
plt.axvline(valor_critico_z_inferior, color='red', linestyle='--', label=f'Valores Críticos Z ({valor_critico_z_inferior:.2f}, {valor_critico_z_superior:.2f})')
plt.axvline(valor_critico_z_superior, color='red', linestyle='--') # No repetir la etiqueta

# Estadístico Z calculado
plt.axvline(estadistico_z, color='blue', linestyle='-', label=f'Estadístico Z Calculado ({estadistico_z:.2f})')

plt.title('Distribución Normal Estándar con Región de Rechazo (Prueba Bilateral)')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
print("-" * 30)

# 7. Tomar una decisión estadística y concluir en términos prácticos, explicando el significado del valor p.
print("7. Decisión estadística y conclusión:")

# Decisión basada en la región de rechazo
if estadistico_z < valor_critico_z_inferior or estadistico_z > valor_critico_z_superior:
    print(f"El estadístico Z calculado ({estadistico_z:.4f}) cae en la región de rechazo.")
    print("Se rechaza la hipótesis nula (H0).")
else:
    print(f"El estadístico Z calculado ({estadistico_z:.4f}) no cae en la región de rechazo.")
    print("No se rechaza la hipótesis nula (H0).")

# Decisión basada en el valor p
print("\nDecisión basada en el valor p:")
if valor_p < nivel_significancia:
    print(f"El valor p ({valor_p:.4f}) es menor que el nivel de significancia ({nivel_significancia}).")
    print("Se rechaza la hipótesis nula (H0).")
else:
    print(f"El valor p ({valor_p:.4f}) no es menor que el nivel de significancia ({nivel_significancia}).")
    print("No se rechaza la hipótesis nula (H0).")

# Concluir en términos prácticos
print("\nConclusión en términos prácticos:")
if valor_p < nivel_significancia:
    print(f"Con un nivel de significancia del {nivel_significancia*100:.0f}%, hay suficiente evidencia estadística para concluir que la media poblacional del volumen de llenado de las botellas es significativamente diferente de 500 mL.")
    print("Esto sugiere que la máquina podría estar descalibrada.")
else:
    print(f"Con un nivel de significancia del {nivel_significancia*100:.0f}%, no hay suficiente evidencia estadística para concluir que la media poblacional del volumen de llenado de las botellas es significativamente diferente de 500 mL.")
    print("No se puede afirmar con esta muestra que la máquina esté descalibrada.")

print(f"\nEl valor p de {valor_p:.4f} representa la probabilidad de obtener una media muestral de 498 mL (o una tan o más extrema en cualquier dirección) si la verdadera media poblacional fuera 500 mL. Dado que este valor es mayor que α = {nivel_significancia}, consideramos que este resultado no es lo suficientemente improbable si la hipótesis nula fuera cierta, lo que nos lleva a no rechazar H0.")
````

**26. Un restaurante afirma que su tiempo promedio de entrega a domicilio es 25 minutos. Un cliente
sospecha que es mayor. Se registran 15 entregas, obteniendo una media de 28 minutos y desviación
estándar muestral s = 4 minutos. Con un nivel de significancia de α = 0,05.**


```
1. Plantee las hipótesis nula y alternativa.
2. Calcule el estadístico t.
3. Determine los grados de libertad y el valor crítico, además de la región crítica y la región de
rechazo.
4. Concluya si se rechaza H0.
```





````python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt

# Problema: Prueba de hipótesis sobre el tiempo promedio de entrega

# 1. Plantear las hipótesis nula y alternativa.
# El restaurante afirma que el tiempo promedio es 25 minutos (μ = 25).
# El cliente sospecha que es MAYOR (μ > 25). Es una prueba unilateral derecha.
# H0: La media poblacional del tiempo de entrega es igual a 25 minutos (μ = 25)
# H1: La media poblacional del tiempo de entrega es mayor que 25 minutos (μ > 25) - Prueba unilateral derecha

print("1. Hipótesis planteadas:")
print("H0: μ = 25")
print("H1: μ > 25")
print("-" * 30)

# 2. Establecer el nivel de significancia α.
nivel_significancia = 0.05
print(f"2. Nivel de significancia α: {nivel_significancia}")
print("-" * 30)

# Datos del problema
media_muestral = 28
media_poblacional_hipotetizada = 25
desviacion_estandar_muestral = 4
tamano_muestra = 15

# 3. Identificar el estadístico de prueba y su distribución.
# Dado que NO se conoce la desviación estándar poblacional (σ) y el tamaño de la muestra es pequeño (n < 30),
# utilizamos el estadístico t de Student.
print("3. Estadístico de prueba: t de Student")
print("Distribución: Distribución t con grados de libertad = n - 1")
print("-" * 30)

# 4. Determinar los grados de libertad y el valor crítico, además de la región crítica y la región de rechazo.
grados_libertad = tamano_muestra - 1
# Para una prueba unilateral derecha, el valor crítico es el valor de t tal que P(t >= valor_critico) = alpha
valor_critico_t = t.ppf(1 - nivel_significancia, grados_libertad)

print(f"4. Grados de libertad: {grados_libertad}")
print(f"Valor crítico t para α = {nivel_significancia}: {valor_critico_t:.4f}")
print(f"Región de rechazo: t > {valor_critico_t:.4f}")
print("-" * 30)

# 5. Calcule el estadístico de prueba y el valor p.
# Calcular el error estándar utilizando la desviación estándar muestral
error_estandar = desviacion_estandar_muestral / np.sqrt(tamano_muestra)
# Calcular el estadístico t
estadistico_t = (media_muestral - media_poblacional_hipotetizada) / error_estandar

print(f"5. Estadístico t calculado: {estadistico_t:.4f}")

# Para una prueba unilateral derecha, el valor p es P(t >= estadistico_t)
valor_p = 1 - t.cdf(estadistico_t, grados_libertad)

print(f"Valor p: {valor_p:.4f}")
print("-" * 30)

# 6. Graficar la distribución correspondiente, indicando claramente la región de rechazo, el valor crítico y el estadístico calculado.
plt.figure(figsize=(10, 6))
# Generar valores para el eje x basados en la distribución t
x = np.linspace(t.ppf(0.001, grados_libertad), t.ppf(0.999, grados_libertad), 1000)
plt.plot(x, t.pdf(x, grados_libertad), label=f'Distribución t ({grados_libertad} GL)')

# Región de rechazo (cola derecha)
x_rechazo = np.linspace(valor_critico_t, x[-1], 100)
plt.fill_between(x_rechazo, 0, t.pdf(x_rechazo, grados_libertad), color='red', alpha=0.5, label=f'Región de Rechazo (α = {nivel_significancia})')

# Valor crítico
plt.axvline(valor_critico_t, color='red', linestyle='--', label=f'Valor Crítico t ({valor_critico_t:.2f})')

# Estadístico t calculado
plt.axvline(estadistico_t, color='blue', linestyle='-', label=f'Estadístico t Calculado ({estadistico_t:.2f})')

plt.title(f'Distribución t ({grados_libertad} GL) con Región de Rechazo (Prueba Unilateral Derecha)')
plt.xlabel('Valor t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
print("-" * 30)


# 7. Tomar una decisión estadística y concluir en términos prácticos, explicando el significado del valor p.
print("7. Decisión estadística y conclusión:")

# Decisión basada en la región de rechazo
if estadistico_t > valor_critico_t:
    print(f"El estadístico t calculado ({estadistico_t:.4f}) cae en la región de rechazo.")
    print("Se rechaza la hipótesis nula (H0).")
else:
    print(f"El estadístico t calculado ({estadistico_t:.4f}) no cae en la región de rechazo.")
    print("No se rechaza la hipótesis nula (H0).")

# Decisión basada en el valor p
print("\nDecisión basada en el valor p:")
if valor_p < nivel_significancia:
    print(f"El valor p ({valor_p:.4f}) es menor que el nivel de significancia ({nivel_significancia}).")
    print("Se rechaza la hipótesis nula (H0).")
else:
    print(f"El valor p ({valor_p:.4f}) no es menor que el nivel de significancia ({nivel_significancia}).")
    print("No se rechaza la hipótesis nula (H0).")

# Concluir en términos prácticos
print("\nConclusión en términos prácticos:")
if valor_p < nivel_significancia:
    print(f"Con un nivel de significancia del {nivel_significancia*100:.0f}%, hay suficiente evidencia estadística para concluir que el tiempo promedio de entrega a domicilio del restaurante es significativamente mayor a 25 minutos.")
    print("Esto respalda la sospecha del cliente.")
else:
    print(f"Con un nivel de significancia del {nivel_significancia*100:.0f}%, no hay suficiente evidencia estadística para concluir que el tiempo promedio de entrega a domicilio del restaurante es significativamente mayor a 25 minutos.")
    print("No se puede afirmar con esta muestra que el tiempo promedio sea mayor a 25 minutos.")

print(f"\nEl valor p de {valor_p:.4f} representa la probabilidad de obtener una media muestral de 28 minutos (o una más extrema) si el verdadero tiempo promedio poblacional fuera 25 minutos. Dado que este valor es mayor que α = {nivel_significancia}, consideramos que este resultado no es lo suficientemente improbable si la hipótesis nula fuera cierta, lo que nos lleva a no rechazar H0.")
````

````python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt

# Problema: Prueba de hipótesis sobre el consumo de combustible de un auto

# 1. Plantear las hipótesis nula y alternativa.
# El fabricante afirma que el consumo promedio es 15 km/L (μ = 15). El problema no especifica si se sospecha que es mayor o menor,
# por lo tanto, asumimos que se quiere verificar si es diferente de 15. Es una prueba bilateral.
# H0: La media poblacional del consumo de combustible es igual a 15 km/L (μ = 15)
# H1: La media poblacional del consumo de combustible es diferente de 15 km/L (μ ≠ 15) - Prueba bilateral

print("1. Hipótesis planteadas:")
print("H0: μ = 15")
print("H1: μ ≠ 15")
print("-" * 30)

# 2. Establecer el nivel de significancia α.
nivel_significancia = 0.05
print(f"2. Nivel de significancia α: {nivel_significancia}")
print("-" * 30)

# Datos del problema
media_muestral = 14
media_poblacional_hipotetizada = 15
desviacion_estandar_muestral = 1.2
tamano_muestra = 10

# 3. Identificar el estadístico de prueba y su distribución.
# Dado que NO se conoce la desviación estándar poblacional (σ) y el tamaño de la muestra es pequeño (n < 30),
# utilizamos el estadístico t de Student.
print("3. Estadístico de prueba: t de Student")
print("Distribución: Distribución t con grados de libertad = n - 1")
print("-" * 30)

# 4. Determinar los grados de libertad y el valor crítico, además de la región crítica y la región de rechazo.
grados_libertad = tamano_muestra - 1
# Para una prueba bilateral con nivel de significancia α, hay dos valores críticos:
# uno negativo (-t_alpha/2) y uno positivo (+t_alpha/2).
# El área en cada cola de la distribución es α/2.
alpha_dos_colas = nivel_significancia / 2
valor_critico_t_inferior = t.ppf(alpha_dos_colas, grados_libertad) # Valor crítico izquierdo
valor_critico_t_superior = t.ppf(1 - alpha_dos_colas, grados_libertad) # Valor crítico derecho

print(f"4. Grados de libertad: {grados_libertad}")
print(f"Valores críticos t para α = {nivel_significancia}: {valor_critico_t_inferior:.4f} y {valor_critico_t_superior:.4f}")
print(f"Región de rechazo: t < {valor_critico_t_inferior:.4f} o t > {valor_critico_t_superior:.4f}")
print("-" * 30)

# 5. Calcule el estadístico de prueba y el valor p.
# Calcular el error estándar utilizando la desviación estándar muestral
error_estandar = desviacion_estandar_muestral / np.sqrt(tamano_muestra)
# Calcular el estadístico t
estadistico_t = (media_muestral - media_poblacional_hipotetizada) / error_estandar

print(f"5. Estadístico t calculado: {estadistico_t:.4f}")

# Para una prueba bilateral, el valor p es el doble del área en la cola más allá del estadístico t calculado.
# Usamos abs() para obtener el área en la cola correcta (izquierda si t es negativo, derecha si t es positivo)
# y luego multiplicamos por 2.
valor_p = 2 * t.cdf(-abs(estadistico_t), grados_libertad) # Área en la cola izquierda de -|t| multiplicada por 2

print(f"Valor p: {valor_p:.4f}")
print("-" * 30)

# 6. Graficar la distribución correspondiente, indicando claramente la región de rechazo, el valor crítico y el estadístico calculado.
plt.figure(figsize=(10, 6))
# Generar valores para el eje x basados en la distribución t, cubriendo un rango adecuado
x = np.linspace(t.ppf(0.001, grados_libertad), t.ppf(0.999, grados_libertad), 1000)
plt.plot(x, t.pdf(x, grados_libertad), label=f'Distribución t ({grados_libertad} GL)')

# Región de rechazo (dos colas)
x_rechazo_inferior = np.linspace(x[0], valor_critico_t_inferior, 100)
plt.fill_between(x_rechazo_inferior, 0, t.pdf(x_rechazo_inferior, grados_libertad), color='red', alpha=0.5, label=f'Región de Rechazo (α/2 = {alpha_dos_colas})')

x_rechazo_superior = np.linspace(valor_critico_t_superior, x[-1], 100)
plt.fill_between(x_rechazo_superior, 0, t.pdf(x_rechazo_superior, grados_libertad), color='red', alpha=0.5) # No repetir la etiqueta

# Valores críticos
plt.axvline(valor_critico_t_inferior, color='red', linestyle='--', label=f'Valores Críticos t ({valor_critico_t_inferior:.2f}, {valor_critico_t_superior:.2f})')
plt.axvline(valor_critico_t_superior, color='red', linestyle='--') # No repetir la etiqueta

# Estadístico t calculado
plt.axvline(estadistico_t, color='blue', linestyle='-', label=f'Estadístico t Calculado ({estadistico_t:.2f})')

plt.title(f'Distribución t ({grados_libertad} GL) con Región de Rechazo (Prueba Bilateral)')
plt.xlabel('Valor t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
print("-" * 30)


# 7. Tomar una decisión estadística y concluir en términos prácticos, explicando el significado del valor p.
print("7. Decisión estadística y conclusión:")

# Decisión basada en la región de rechazo
if estadistico_t < valor_critico_t_inferior or estadistico_t > valor_critico_t_superior:
    print(f"El estadístico t calculado ({estadistico_t:.4f}) cae en la región de rechazo.")
    print("Se rechaza la hipótesis nula (H0).")
else:
    print(f"El estadístico t calculado ({estadistico_t:.4f}) no cae en la región de rechazo.")
    print("No se rechaza la hipótesis nula (H0).")

# Decisión basada en el valor p
print("\nDecisión basada en el valor p:")
if valor_p < nivel_significancia:
    print(f"El valor p ({valor_p:.4f}) es menor que el nivel de significancia ({nivel_significancia}).")
    print("Se rechaza la hipótesis nula (H0).")
else:
    print(f"El valor p ({valor_p:.4f}) no es menor que el nivel de significancia ({nivel_significancia}).")
    print("No se rechaza la hipótesis nula (H0).")

# Concluir en términos prácticos
print("\nConclusión en términos prácticos:")
if valor_p < nivel_significancia:
    print(f"Con un nivel de significancia del {nivel_significancia*100:.0f}%, hay suficiente evidencia estadística para concluir que el consumo promedio poblacional del auto es significativamente diferente de 15 km/L.")
else:
    print(f"Con un nivel de significancia del {nivel_significancia*100:.0f}%, no hay suficiente evidencia estadística para concluir que el consumo promedio poblacional del auto es significativamente diferente de 15 km/L.")
    print("No se puede afirmar con esta muestra que el consumo promedio sea diferente de 15 km/L.")

print(f"\nEl valor p de {valor_p:.4f} representa la probabilidad de obtener una media muestral de 14 km/L (o una tan o más extrema en cualquier dirección) si la verdadera media poblacional fuera 15 km/L. Dado que este valor es mayor que α = {nivel_significancia}, consideramos que este resultado no es lo suficientemente improbable si la hipótesis nula fuera cierta, lo que nos lleva a no rechazar H0.")
````

````python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# Problema: Prueba de hipótesis sobre la varianza del tiempo de duración de las llamadas

# 1. Plantear las hipótesis nula y alternativa.
# La compañía afirma que la varianza poblacional es 2.5 (σ^2 = 2.5). Se pregunta si hay evidencia de que la varianza poblacional difiere de 2.5.
# Como se pregunta si DIFIERE (puede ser mayor o menor), es una prueba bilateral.
# H0: La varianza poblacional del tiempo de duración de las llamadas es igual a 2.5 minutos² (σ² = 2.5)
# H1: La varianza poblacional del tiempo de duración de las llamadas es diferente de 2.5 minutos² (σ² ≠ 2.5) - Prueba bilateral

print("1. Hipótesis planteadas:")
print("H0: σ² = 2.5")
print("H1: σ² ≠ 2.5")
print("-" * 30)

# 2. Establecer el nivel de significancia α.
nivel_significancia = 0.05
print(f"2. Nivel de significancia α: {nivel_significancia}")
print("-" * 30)

# Datos del problema
varianza_muestral = 4
varianza_poblacional_hipotetizada = 2.5
tamano_muestra = 16

# 3. Identificar el estadístico de prueba y su distribución.
# Para probar hipótesis sobre la varianza poblacional con la varianza muestral, utilizamos el estadístico Chi-cuadrado (χ²)
# y su distribución.
print("3. Estadístico de prueba: Chi-cuadrado (χ²)")
print("Distribución: Distribución Chi-cuadrado con grados de libertad = n - 1")
print("-" * 30)

# 4. Calcule el estadístico Chi-cuadrado.
# Fórmula: χ² = ((n - 1) * s²) / σ²
estadistico_chi2 = ((tamano_muestra - 1) * varianza_muestral) / varianza_poblacional_hipotetizada

print(f"4. Estadístico χ² calculado: {estadistico_chi2:.4f}")
print("-" * 30)


# 5. Determine los grados de libertad y los valores críticos, además de la región crítica y la región de rechazo.
grados_libertad = tamano_muestra - 1
# Para una prueba bilateral con nivel de significancia α, hay dos valores críticos:
# uno inferior y uno superior.
# El área en cada cola de la distribución es α/2.
alpha_dos_colas = nivel_significancia / 2
valor_critico_chi2_inferior = chi2.ppf(alpha_dos_colas, grados_libertad) # Valor crítico izquierdo
valor_critico_chi2_superior = chi2.ppf(1 - alpha_dos_colas, grados_libertad) # Valor crítico derecho

print(f"5. Grados de libertad: {grados_libertad}")
print(f"Valores críticos χ² para α = {nivel_significancia}: {valor_critico_chi2_inferior:.4f} (inferior) y {valor_critico_chi2_superior:.4f} (superior)")
print(f"Región de rechazo: χ² < {valor_critico_chi2_inferior:.4f} o χ² > {valor_critico_chi2_superior:.4f}")
print("-" * 30)

# 6. Calcular el valor p.
# Para una prueba bilateral, el valor p es el doble del área en la cola más allá del estadístico Chi-cuadrado calculado.
# Si el estadístico Chi-cuadrado es menor que la media de la distribución (que es igual a los grados de libertad),
# calculamos el área a la izquierda y la multiplicamos por 2.
# Si el estadístico Chi-cuadrado es mayor que la media, calculamos el área a la derecha y la multiplicamos por 2.
# Una forma más general es calcular el área de la cola "más cercana" al estadístico y duplicarla.
if estadistico_chi2 < grados_libertad: # Si el estadístico está a la izquierda de la media
    valor_p = 2 * chi2.cdf(estadistico_chi2, grados_libertad)
else: # Si el estadistico está a la derecha de la media
    valor_p = 2 * (1 - chi2.cdf(estadistico_chi2, grados_libertad))

print(f"Valor p: {valor_p:.4f}")
print("-" * 30)

# 7. Graficar la distribución correspondiente, indicando claramente la región de rechazo, los valores críticos y el estadístico calculado.
plt.figure(figsize=(10, 6))
# Generar valores para el eje x basados en la distribución chi-cuadrado
# Ajustar el rango para asegurar que los valores críticos y el estadístico calculado sean visibles
x = np.linspace(chi2.ppf(0.0001, grados_libertad), chi2.ppf(0.9999, grados_libertad), 1000)
plt.plot(x, chi2.pdf(x, grados_libertad), label=f'Distribución χ² ({grados_libertad} GL)')

# Región de rechazo (dos colas)
x_rechazo_inferior = np.linspace(x[0], valor_critico_chi2_inferior, 100)
plt.fill_between(x_rechazo_inferior, 0, chi2.pdf(x_rechazo_inferior, grados_libertad), color='red', alpha=0.5, label=f'Región de Rechazo (α/2 = {alpha_dos_colas})')

x_rechazo_superior = np.linspace(valor_critico_chi2_superior, x[-1], 100)
plt.fill_between(x_rechazo_superior, 0, chi2.pdf(x_rechazo_superior, grados_libertad), color='red', alpha=0.5) # No repetir la etiqueta

# Valores críticos
plt.axvline(valor_critico_chi2_inferior, color='red', linestyle='--', label=f'Valores Críticos χ² ({valor_critico_chi2_inferior:.2f}, {valor_critico_chi2_superior:.2f})')
plt.axvline(valor_critico_chi2_superior, color='red', linestyle='--') # No repetir la etiqueta

# Estadístico Chi-cuadrado calculado
plt.axvline(estadistico_chi2, color='blue', linestyle='-', label=f'Estadístico χ² Calculado ({estadistico_chi2:.2f})')

plt.title(f'Distribución χ² ({grados_libertad} GL) con Región de Rechazo (Prueba Bilateral)')
plt.xlabel('Valor χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
print("-" * 30)


# 8. Tomar una decisión estadística y concluir en términos prácticos, explicando el significado del valor p.
print("8. Decisión estadística y conclusión:")

# Decisión basada en la región de rechazo
if estadistico_chi2 < valor_critico_chi2_inferior or estadistico_chi2 > valor_critico_chi2_superior:
    print(f"El estadístico χ² calculado ({estadistico_chi2:.4f}) cae en la región de rechazo.")
    print("Se rechaza la hipótesis nula (H0).")
else:
    print(f"El estadístico χ² calculado ({estadistico_chi2:.4f}) no cae en la región de rechazo.")
    print("No se rechaza la hipótesis nula (H0).")

# Decisión basada en el valor p
print("\nDecisión basada en el valor p:")
if valor_p < nivel_significancia:
    print(f"El valor p ({valor_p:.4f}) es menor que el nivel de significancia ({nivel_significancia}).")
    print("Se rechaza la hipótesis nula (H0).")
else:
    print(f"El valor p ({valor_p:.4f}) no es menor que el nivel de significancia ({nivel_significancia}).")
    print("No se rechaza la hipótesis nula (H0).")

# Concluir en términos prácticos
print("\nConclusión en términos prácticos:")
if valor_p < nivel_significancia:
    print(f"Con un nivel de significancia del {nivel_significancia*100:.0f}%, hay suficiente evidencia estadística para concluir que la varianza poblacional del tiempo de duración de las llamadas es significativamente diferente de 2.5 minutos².")
else:
    print(f"Con un nivel de significancia del {nivel_significancia*100:.0f}%, no hay suficiente evidencia estadística para concluir que la varianza poblacional del tiempo de duración de las llamadas es significativamente diferente de 2.5 minutos².")
    print("No se puede afirmar con esta muestra que la varianza poblacional sea diferente de 2.5 minutos².")

print(f"\nEl valor p de {valor_p:.4f} representa la probabilidad de obtener una varianza muestral de 4 minutos² (o una tan o más extrema en cualquier dirección) si la verdadera varianza poblacional fuera 2.5 minutos². Dado que este valor es mayor que α = {nivel_significancia}, consideramos que este resultado no es lo suficientemente improbable si la hipótesis nula fuera cierta, lo que nos lleva a no rechazar H0.")
````
