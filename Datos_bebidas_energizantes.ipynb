```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar archivo CSV
df = pd.read_excel("bebidas.xlsx")

# Información básica
print("Información del dataset")
print(df.info())
print("Primeras filas del dataset")
print(df.head())
print("Resumen estadístico del dataset")
print(df.describe())
```

```python

# Valores faltantes
print("\nValores faltantes por columna:")
print(df.isnull().sum())

# Visualización de nulos
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Mapa de datos faltantes")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograma para Edad
plt.figure(figsize=(8, 6))
sns.histplot(df['2. ¿En qué rango de edad se encuentra usted?'], bins=10, kde=False)
plt.xlabel('Rango de Edad')
plt.ylabel('Frecuencia')
plt.title('Histograma de Rango de Edad')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Diagrama de caja para Edad
plt.figure(figsize=(6, 6))
sns.boxplot(y=df['2. ¿En qué rango de edad se encuentra usted?'])
plt.title('Diagrama de Caja - Rango de Edad')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Diagrama de caja para Semestre
plt.figure(figsize=(6, 6))
sns.boxplot(y=df[' 5. ¿En que semestre se encuentra usted?'])
plt.title('Diagrama de Caja - Semestre')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Lista de variables categóricas y títulos para los gráficos
categorical_vars = [
    ('13. ¿Desde hace cuánto tiempo consume bebidas energizantes?', 'Distribución por consumo de bebidas energizantes')
]

for var, title in categorical_vars:
    plt.figure(figsize=(8, 8))
    df[var].value_counts().plot.pie(
        autopct='%1.1f%%',
        startangle=90,
        colors=plt.cm.tab20.colors
    )
    plt.title(title)
    plt.ylabel('')  # Eliminar la etiqueta del eje y
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt

# Lista de variables categóricas y títulos para los gráficos
categorical_vars = [
    ('12. ¿Con qué frecuencia consume bebidas energizantes?', 'Distribución por frecuencia de consumo de bebidas energizantes'),
]

for var, title in categorical_vars:
    plt.figure(figsize=(8, 8))
    df[var].value_counts().plot.pie(
        autopct='%1.1f%%',
        startangle=90,
        colors=plt.cm.tab20.colors
    )
    plt.title(title)
    plt.ylabel('')  # Eliminar la etiqueta del eje y
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Lista de variables categóricas y títulos para los gráficos de barras
categorical_vars_bar = [
    ('14. ¿Por qué consume bebidas energizantes? (puede marcar varias opciones)', 'Distribución por el motivo de consumir bebidas energizantes')
]

for var, title in categorical_vars_bar:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=var, palette='viridis')
    plt.title(title)
    plt.xlabel(var)
    plt.ylabel('Frecuencia')
    plt.xticks(rotation=45, ha='right')  # Mejora la legibilidad de las etiquetas
    plt.tight_layout()
    plt.show()
```
